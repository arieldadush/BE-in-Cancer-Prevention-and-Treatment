```python
# merge by left joining - the S3 database with ilana VCF with freq
merge_df = pd.merge(edit_snv_icgc_df, hg19_genes_and_strand_df_selected_col_without_dup_real_genes, how='left', left_on=['gene',], right_on=['name2',])
merge_df
```

```python
merge_df = merge_df[~merge_df['#chrom'].isna()]
```

```python
# Define your dictionary for character replacement
char_dict = {'C': 'G', 'G': 'C', 'A': 'T', 'T': 'A'}

# Apply the character replacement using a lambda function with a condition
merge_df['stranded_ref'] = merge_df.apply(lambda row: ''.join([char_dict.get(char, char) if row['strand'] == '-' else char for char in row['ref']]), axis=1)
merge_df['stranded_alt'] = merge_df.apply(lambda row: ''.join([char_dict.get(char, char) if row['strand'] == '-' else char for char in row['alt']]), axis=1)

merge_df
```

```python
# Get the index of the columns
index1 = merge_df.columns.get_loc('gene')
index2 = merge_df.columns.get_loc('chrom')

# Swap the columns
merge_df.iloc[:, [index1, index2]] = merge_df.iloc[:, [index2, index1]]
merge_df.rename(columns={'gene': 'chrom', 'chrom': 'gene'}, inplace=True)
merge_df
```

```python
# Get the index of the columns
index1 = merge_df.columns.get_loc('pos')
index2 = merge_df.columns.get_loc('ttype')

# Swap the columns
merge_df.iloc[:, [index1, index2]] = merge_df.iloc[:, [index2, index1]]
merge_df.rename(columns={'pos': 'ttype', 'ttype': 'pos'}, inplace=True)
```

```python
# Get the index of the columns
index1 = merge_df.columns.get_loc('stranded_ref')
index2 = merge_df.columns.get_loc('chr')

# Swap the columns
merge_df.iloc[:, [index1, index2]] = merge_df.iloc[:, [index2, index1]]
merge_df.rename(columns={'stranded_ref': 'chr', 'chr': 'stranded_ref'}, inplace=True)
```

```python
# Get the index of the columns
index1 = merge_df.columns.get_loc('stranded_alt')
index2 = merge_df.columns.get_loc('ttype')

# Swap the columns
merge_df.iloc[:, [index1, index2]] = merge_df.iloc[:, [index2, index1]]
merge_df.rename(columns={'stranded_alt': 'ttype', 'ttype': 'stranded_alt'}, inplace=True)
```

```python
merge_df
```

```python
# Get the index of the columns
index1 = merge_df.columns.get_loc('strand')
index2 = merge_df.columns.get_loc('ref')

# Swap the columns
merge_df.iloc[:, [index1, index2]] = merge_df.iloc[:, [index2, index1]]
merge_df.rename(columns={'strand': 'ref', 'ref': 'strand'}, inplace=True)
```

```python
merge_df['start'] = merge_df['pos']-1
```

```python
# Get the index of the columns
index1 = merge_df.columns.get_loc('start')
index2 = merge_df.columns.get_loc('sample_id')

# Swap the columns
merge_df.iloc[:, [index1, index2]] = merge_df.iloc[:, [index2, index1]]
merge_df.rename(columns={'start': 'sample_id', 'sample_id': 'start'}, inplace=True)
```
